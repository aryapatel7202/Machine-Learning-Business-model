```python
import pandas as pd
import numpy as pn
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df = pd.read_csv(r'1000_Companies.csv')
df
```

```python
X = df[['Administration']]
y = df['Profit']
```

```python
X_tr,X_te,y_tr,y_te = train_test_split(X,y,test_size = 0.2,random_state=2)
model = LinearRegression()
```

```python
model.fit(X_tr,y_tr)
```

```python
y_pred = model.predict(X_te)
```

```python
print({'actual':[y_te],'pred':[y_pred]})
```

```python
plt.scatter(X_te,y_te,color='blue',label='actual vs perdicted')
plt.plot(X_te,y_pred,color='r',label="perfect prediction")
plt.legend()
plt.show()
```
